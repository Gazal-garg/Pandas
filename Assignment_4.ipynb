```python
import pandas as pd
df=pd.read_csv(r"C:\Users\Gazal\Documents\matches.csv")
df1=pd.read_csv(r"C:\Users\Gazal\Downloads\deliveries.csv")
mask=df1['player_dismissed'].isnull()
wictkn=df1[~mask]
tbol=wictkn.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).index[0:10]
tbol_new=tbol.tolist() 
wictkn[wictkn['bowler'].isin(tbol_new)].pivot_table(index = "bowler", columns = "over", values = "player_dismissed", aggfunc = "count") 

```

```python
# Q2.- Make a function for batting profile of any batsman:-                                              [Medium+]
```

Ex- 

V Kohli:

Innings:  

Total Runs: 

No. of MOM(Man of the match):

No. of boundries:

```python
import pandas as pd
df1=pd.read_csv(r"C:\Users\Gazal\Downloads\deliveries.csv")
df=pd.read_csv(r"C:\Users\Gazal\Documents\matches.csv")
def batting_profile(bman):
    print("Profile of ",bman)
    d=df1.groupby('batsman')
    i=len(d.get_group(bman)['match_id'].unique())
    print("Innings:",i)
    tr=d.get_group(bman)['total_runs'].sum()
    print("Total runs:",tr)
    mn=(df['player_of_match']==bman).sum()
    print("No of Man of Match: ",mn)
    bo=(df1['batsman_runs']>3).sum()
    print("No of boundaries",bo)
batting_profile('SR Watson')

    



```

```python
# Q3.- Write a function to return top-4 batsman at a venue. Take venue as input.
```

```python
import pandas as pd
df=pd.read_csv(r"C:\Users\Gazal\Documents\matches.csv")
df1=pd.read_csv(r"C:\Users\Gazal\Downloads\deliveries.csv")
def cricket(ven):
    new=df1.merge(df,left_on='match_id',right_on='id')  
    b=new['venue']==ven
    a=new[b].groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).index[0:4]
    print("Top-4 batsman player at",ven,":")
    for i in a:
        print(i)
        
    
cricket("Eden Gardens")
```
